```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read csv using pandas
data = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/diamond.csv')

# check value counts of Cut column
data['Cut'].value_counts()
```

```python
# Scatter plot of prices vs. clarity
data.plot(x='Clarity', y='Price', kind='scatter', title="Prices by clarity")

# Show the plot
plt.show()
```

```python
# Bar Chart - Top 10 Product Categories by Sales
top_10_categories = data.groupby('Clarity')['Price'].sum().nlargest(10)
plt.bar(top_10_categories.index, top_10_categories.values)
plt.xticks(rotation=45)
plt.xlabel('Clarity')
plt.ylabel('Price')
plt.title('Top 10 Clarity by price')
plt.show()
```

```python
# Show correlation between price and Carat weight
data.corr()
```

```python
# Visualize the relation using heatmap
sns.heatmap(data.corr(), cmap="YlGnBu", annot=True)
```

```python
# Box Plot - Prices by Carat Weight
plt.figure(figsize=(8, 6)) 
sns.boxplot(x='Carat Weight', y='Price', data=data) 
plt.xlabel('Carat Weight') 
plt.ylabel('Price') 
plt.title('Prices by Carat Weight') 
plt.show()
```

```python
# Violin Plot - Price Distribution by Clarity 
plt.figure(figsize=(10, 6)) 
sns.violinplot(x='Clarity', y='Price', data=data) 
plt.xlabel('Clarity') 
plt.ylabel('Price') 
plt.title('Price Distribution by Clarity') 
plt.show()
```
